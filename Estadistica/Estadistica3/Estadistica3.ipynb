```python
#Importando Librerias
import DistribucionFrecuencias as df
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
```

```python
archivo = open('/home/jesus/clases/Estadistica3/medidas.csv')
data = []

for i in archivo:
    data.append(float(i))

archivo.close()
```

```python
N = df.NumeroDatos(data)
maximo = max(data)
minimo = min(data)
R = df.Rango(maximo,minimo)
K = df.NumeroClases(N)
I = df.IntervaloClases(R,K)
limites = df.Limites(I,minimo,K)
frecuencia = df.Contar(data,limites,I,maximo)
marca_clase = df.MarcaClase(limites)
intervalos = df.Intervalos(limites)
h = df.FrecuenciaRelativa(frecuencia,N)
F = df.FrecuenciaAcumulada(frecuencia)
H = df.FrecuenciaRelativaAcumulada(F,N)
```

```python
plot.hist(x=data,bins=K,rwidth=0.85,color='#4800FF')
plot.xticks(limites) #Numero divisiones en el eje x.

print(limites)
eje_y = []
for i in range(0,max(frecuencia)+1):
    if i%2 == 0:
        eje_y.append(i)

plot.yticks(eje_y)
plot.xlabel("Intervalos") #Leyenda en el eje x
plot.ylabel("Frecuencia") #Leyenda en el eje y
plot.title('Histograma y poligono de "frecuencias"') #Titulo de la grafica

plot.plot(marca_clase,frecuencia,color="#000000",marker="o")


```
