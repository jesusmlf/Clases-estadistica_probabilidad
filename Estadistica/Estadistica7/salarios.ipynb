```python
import pandas as pd
import matplotlib.pyplot as plot
from pearson import Pearson
from Regresion import ordenada_origen,pendiente

data = pd.read_csv("data/salarios-pedagogo-experiencia (1).csv")
experiencia = data["experiencia"]
salario = data["salario"]

plot.scatter(experiencia,salario,color="green",marker="^")

coeficiente_pearson = Pearson(experiencia,salario)
print(coeficiente_pearson)

m = pendiente(experiencia,salario)
b = ordenada_origen(experiencia,salario,m)

def f(x,m,b):
    return (m*x)+b

x = range(0,10)
plot.plot (x, [f(i,m,b) for i in x],color="black")

print(f(2.5,m,b))

plot.xlabel("Experiencia (a√±os)")
plot.ylabel("Salario (Pesos mx)")
plot.title("Salarios Pedagogos - Experiencia")
plot.show()
```
